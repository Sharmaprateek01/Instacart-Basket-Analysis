```python
#Creating a new column named max_order
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
##Creating loyalty flag 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge ['loyalty_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(50)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','max','min']})
```

```python
#Performing single agg to calculate mean of prices grouped by user_id
ords_prods_merge.groupby('user_id').agg({'prices':['mean']})
```

```python
#Creating a new column named spending_habits
ords_prods_merge['spending_habits'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
##Creating spending flag 
ords_prods_merge.loc[ords_prods_merge['spending_habits'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spending_habits'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge ['spending_flag'].value_counts(dropna= False)
```

```python
#Performing single agg to calculate mean of prices grouped by user_id
ords_prods_merge.groupby('user_id').agg({'days_since_prior_order':['median']})
```

```python
#Creating a new column named order_frequency
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Creating order frequency flag
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge ['order_frequency_flag'].value_counts(dropna= False)
```
