```python
df = df_merged[:1000000]
```

```python
#Creating and defining a user-defined function with If-statements
def price_label(row):

    if row['prices']<= 5:
        return'Low-range product'
    elif (row['prices']> 5)and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices']> 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
df['price_range']= df.apply(price_label, axis=1)
```

```python
#Checking the values
df['price_range'].value_counts(dropna=False)
```

```python
#to check the most expensive product
df['prices'].max()
```

Creating and defining with loc () function with If-statements

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Checking the values
df['price_range_loc'].value_counts(dropna= False)
```

```python
#Creating new column Busiest days and slowest days:
result1 = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0 or value ==1:
    result1.append("Busiest days")
  elif value == 4 or value == 3:
    result1.append("Slowest days")
  else:
    result1.append("Regularly busy")
```

Categorizing value counts above into 3 categories:

```python
#Creating If statement with For-loops
result2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    result2.append("Most orders")
  elif value in [3,4,2,5,1,0,6,23]:
    result2.append("Fewest orders")
  else:
    result2.append("Average orders")
```
